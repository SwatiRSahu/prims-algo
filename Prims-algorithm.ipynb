```python
import numpy as np


def startPrismAlgo():
    for x in range(1, 6):
        INF = 9999999
        # number of nodes in graph
        choice = input("Want to find minimum spanning tree?(Y/N): ")
        if choice == "Y" or choice == "y":
            num = int(input("enter number of vertices:"))
            N = num

            if N > 1:
                # creating graph by adjacency matrix method
                G = np.empty([N, N], dtype=int)
                # To fill adjacency matrix with 0
                for i in G:
                    i.fill(0)

                # To assign random weigths to paths
                for i in range(N):
                    for j in range(N):
                        # To check if path is a self loop
                        if i == j:
                            G[i][j] = 0
                        # To check if weight is assigned to path j to i and i to j
                        elif G[i][j] == 0 and G[j][i] == 0:
                            G[i][j] = np.random.randint(low=0, high=10, size=1)[0]
                            G[j][i] = G[i][j]

                selected_node = []
                for i in range(N):
                    selected_node.append(0)

                no_edge = 0

                selected_node[0] = True

                # Pinting adjacency matrix G
                print()
                print("Matrix representation of", x, " graph")
                for i in range(N):
                    print("\t|%4d" % (i), end="")
                print()

                for i in range(55):
                    print("-", end="")
                print()

                for i in range(N):
                    print("%4d" % i, end="")
                    for j in range(N):
                        print("\t|%4d" % G[i][j], end="")
                    print()
                    for i in range(55):
                        print("-", end="")
                    print()

                # printing for edge and weight
                print("\nThe minimum Spanning Tree of ", x, "graph by Prim's Algorithm")
                print("\nEdge : Weight")
                while no_edge < N - 1:
                    minimum = INF
                    a = 0
                    b = 0
                    for m in range(N):
                        if selected_node[m]:
                            for n in range(N):
                                if (not selected_node[n]) and G[m][n]:
                                    # not in selected and there is an edge
                                    if minimum > G[m][n]:
                                        minimum = G[m][n]
                                        a = m
                                        b = n
                    print(str(a) + "<->" + str(b) + " : " + str(G[a][b]))
                    selected_node[b] = True
                    no_edge += 1
            else:
                print("Number of nodes is less than 2")
        else:
            break
```

```python
startPrismAlgo()
```
